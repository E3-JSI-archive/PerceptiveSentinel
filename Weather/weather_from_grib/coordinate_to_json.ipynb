```python
import json
import pandas as pd
```

```python
coordinate_list = []
n = 0
for i in range(8):
    for j in range(15):
        coordinate_list.append({
            'lon': 13.25 + j * 0.25,
            'lat': 47 - i * 0.25,
            'id': n
        })
        n = n + 1
```

```python
coordinate_list
```

```python
coor = json.dumps(coordinate_list, indent=2)
```

```python
obj = open('coordinate.json', 'w')
obj.write(coor)
obj.close()
```
